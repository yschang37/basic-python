```python
def applyToEach(L, f):
    """
    assume L is a list, f a function
    mutates L by replacing each element,
    e, of L by f(e)
    """
    for i in range(len(L)):
        L[i] = f(L[i])
```

```python
L = [1, 2, -3, 5.5]
```

```python
applyToEach(L, abs)
```

```python
applyToEach(L, int)
```
