```python
def mult_iter(a,b):
    result = 0
    while b>0:
        result += a
        b -= 1
    return result
```

```python
mult_iter(2,6)
```

```python
# it is really cool because it can compute recurssively

def mult(a,b):
    if b == 1:
        return a
    else:
        return a + mult(a, b-29)
```

```python
mult(3,30)
```

```python
def fact(n):
    if n==1:
        return n
    else:
        return n*fact(n-1)
        
```

```python
fact(7)
```

```python
for i in range(1,10): # it can not be 0, otherwise it will cause the infinite compute
    print(fact(i))
```

```python
def iterPower(base, exp):
    
    result = 1
    while exp>0:
        result=result*base
        exp -= 1
    return result
```

```python
def recurPower(base, exp):
    if exp<=1:
        return base
    else:
        return base*recurPower(base, exp-1)
```
